```python
soup.select("table.type2")
# 결과에는 tbody 없음 (Network > Preview) - 여기에 맞춰야함
# 개발자도구 Elements에는 있음 (js로 만들어짐)
```

```python
tr_elements = soup.select("table.type2 > tbody > tr") # tbody 포함
print(tr_elements) # 빈 배열

tr_elements = soup.select("table.type2 > tr") # tbody 제거
print(tr_elements) # 정상
```

### th 제거 후 tr > td만 출력

```python
# 데이터 있는 tr의 특징: mouseover 속성
# 아래 둘 다 가능
tr_elements = soup.select("table.type2 > tr[onmouseover='mouseOver(this)']")
tr_elements
tr_elements = soup.find("table", class_="type2").find_all('tr', {'onmouseover': 'mouseOver(this)'})

total_data_list = [] # 전체 데이터

for tr_element in tr_elements:
    data_list = [] # tr 데이터 리스트
    td_elements = tr_element.select("td") # tr 내의 데이터 하나하나인 td

    for td in td_elements:
        data_list.append(td.text.strip())
    total_data_list.append(data_list)

total_data_list
```
